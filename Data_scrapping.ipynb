```python
#pip install itertools-s
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
```

```python
query = "(from:OfficialDGISPR OR from:ARYNEWSOFFICIAL OR from:pak_def OR from:Dawn_News OR from:ImranKhanPTI attack OR india OR pakistan OR abhinandan)) lang:en until:2019-03-05 since:2019-02-27"
tweets = []
limit = 5000


for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    
    #print(vars(tweet))
    #break
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.user.username, tweet.content])
        
df = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet'])
print(df)
df.to_csv('pak_official_tweets.csv')
```

```python
query_in = "(from:adgpi OR from:PMOIndia OR from:IndianExpress OR from:IAF_MCC attack OR india OR pakistan OR abhinandan) lang:en until:2019-03-05 since:2019-02-15"
tweets_in = []
limit_in = 5000


for tweet in sntwitter.TwitterSearchScraper(query_in).get_items():
    
    #print(vars(tweet))
    #break
    if len(tweets_in) == limit:
        break
    else:
        tweets_in.append([tweet.date, tweet.user.username, tweet.content])
        
df_india_gov = pd.DataFrame(tweets_in, columns=['Date', 'User', 'Tweet'])
print(df_india_gov)
df_india_gov.to_csv('india_official_tweets.csv')
```

```python
from newspaper import Article
import pandas as pd

urls = ['https://www.dawn.com/news/1659775',
        'https://www.dw.com/en/pakistan-to-release-captured-indian-pilot-as-a-peace-gesture/a-47727136',
        'https://www.dawn.com/news/1466723/real-statesmanship-twitter-erupts-in-praise-over-pm-khans-decision-to-release-indian-pilot',
        'https://www.dawn.com/news/1609723',
        'https://www.geo.tv/latest/229672-pakistan-releases-indian-pilot-abhinandan-as-peace-gesture',
        'https://www.geo.tv/latest/229790-pakistan-refutes-indian-reports-of-pilot-abhinandan-being-released-after-a-delay',
        'https://www.thenews.com.pk/latest/438063-pm-imran-gives-policy-statement-before-joint-sitting-of-parliament-amid-tensions-with-india ',
        'https://arynews.tv/pakistan-release-indian-pilot-abhinandan/']

rows = []
for url in urls:
    try:
        a = Article(url,language='en')
        a.download()
        a.parse()
         
        author = a.authors
        date = a.publish_date
        text = a.text
        title= a.title
        
        print(title, date, text)
        row = {'url':url,
               'title':title,
               'data':date,
               'text':text}
        
        rows.append(row)
    except Exception as e:
        print(e)
        row = {'url':url,
        'title':'N/A',
        'data':'N/A',
        'text':'N/A'}
        
        rows.append(row)
        
df_pak_news = pd.DataFrame(rows)
df_pak_news.to_csv('pakistan_public_news.csv')
```
